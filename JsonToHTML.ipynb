```python
json_data = """
{
   "input":[
      {
         "edition":"4th",
         "name":"JBCN Conference",
         "startDate":"2018-06-11",
         "endDate":"2018-06-13",
         "location":[
            {
               "city":"Barcelona",
               "country":"Spain"
            }
         ]
      },
      {
         "edition":"3rd",
         "name":"DevTernity",
         "startDate":"2018-11-30",
         "endDate":"2018-12-01",
         "location":[
            {
               "city":"Riga",
               "country":"Latvia"
            }
         ]
      },
      {
         "edition":"1st",
         "name":"I T.A.K.E Unconference",
         "startDate":"2016-05-19",
         "endDate":"2016-05-20",
         "location":[
            {
               "city":"Bucharest",
               "country":"Romania"
            },
            {
               "city":"Maramures",
               "country":"Romania"
            }
         ]
      },
      {
         "edition":"2nd",
         "name":"Product Owner Rule Book",
         "startDate":"2016-04-11",
         "endDate":"2016-04-13",
         "location":[
            {
               "city":"Paris",
               "country":"France"
            },
            {
               "city":"Madrid",
               "country":"Spain"
            }
         ]
      },
      {
         "name":"Upfront Summit",
         "startDate":"2018-02-01",
         "location":[
            {
               "city":"Los Angeles",
               "state":"California",
               "country":"United States"
            }
         ]
      },
      {
         "name":"IBM Think",
         "startDate":"2018-03-19",
         "location":[
            {
               "state":"Nevada",
               "country":"United States"
            }
         ]
      }
   ]
}
"""
```

```python
#Imports...
import json
```

```python
def getFancyStructure(json_data):
    
    firstTime = True
    info = ""
    json_load = (json.loads(json_data))
    
    output = """{
    "meetUps": ["""

    data = json_load['input']
    for x in data:
        gotEdition = False

        if firstTime == True:
            firstTime = False
            output += "\n"
        else:
            output += ", \n"
        output += "\t"
        if 'edition' in x:
            gotEdition = True
            output += '"' + (x['edition']) + " "
        if 'name' in x:
            if gotEdition == False:
                output += '"'
            output += (x['name']) + " · "
        if 'startDate' in x:
            output += (x['startDate']) + " "
            if 'endDate' in x:
                output += "/ "
            else:
                output += "· "
        if 'endDate' in x:
            output += (x['endDate']) + " · "

        aux = x['location']
        secondTime = False
        countries = []
        cities = []

        for y in aux:
            if secondTime == True:
                info += " | "
            else:
                hasThree = False
            if 'city' in y and 'state' in y and 'country' in y:
                hasThree = True

            if 'city' in y:
                info += (y['city']) + ", "
                cities.append(y['city'])
            if 'state' in y:
                info += (y['state'])
                if hasThree == False:
                    info += ", "
                else:
                    info += ". "
            if 'country' in y:
                info += (y['country']) 
                if(len(aux) == 2):
                    if(secondTime == False):
                        countries.append(y['country'])
                    else:
                        if(countries[0] == y['country']):
                            #if that's the case, we just print the country a single time (case of Romania).
                            info = cities[0] + " | " + cities[1] + ", " + countries[0]

                if(secondTime == True or len(aux) == 1):
                    info += '"'

            secondTime = True
        output += info
        info = ""


    output += "\n\t]\n}"
    return output
```

---

```python
final_data = getFancyStructure(json_data)
print(final_data)
```

**HOW WAS IT DONE**<br><br>This code is actually split in two parts, the two big <u> for loops</u>. The first one ranges over all the diverse **editions, names, and start and end dates**. Moreover, it adapts to the data available in order to gather all the possible data without causing errors. <br><br>Furthermore, the second for loop goes over all the different **locations** where the events are held. It performs in a very analogous way to the first loop. Nonetheless, it has the particularity that the data gathered is first stored in a variable called 'info' before directly adding that data to the return value. I made it like this owing to the fact that there are some particular cases, like the cities Bucharest and Maramures (which are both from Romania), and consequently, they have to be displayed in a different manner than the usual pair city-country.
